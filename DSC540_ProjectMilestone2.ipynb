```python
import pandas as pd
import numpy as np
```

```python
#importing NFL Passing csv
df = pd.read_csv(r'C:\Users\nickm\Documents\Career_Stats_Passing.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
#removing irrelevant columns
df = df[['Name', 'Position', 'Year', 'Team', 'Passes Attempted', 'Passes Completed', 'Pass Attempts Per Game', 'TD Passes', 'Ints', 'Passer Rating']]
```

```python
df.shape
```

```python
#Removing invalid row values from important fields
df = df[ df['Passes Attempted']!='--']
```

```python
df.shape
```

```python
df.head(20)
```

```python
#Changing the data type of the Passes attempted field to enable filtering 
df['Passes Attempted']= df['Passes Attempted'].astype(int)
```

```python
df.head(20)
```

```python
#creating a data frame for only QB data. 
#Could not simply filter on "Position" column because of Null values. 
#any player with 200 or more attempts will be considered a QB
qb_df= df[ df['Passes Attempted'] > 199]
```

```python
qb_df.shape
```

```python
qb_df.head(20)
```

```python
#filling null values in the 'Position' column
qb_df['Position'] = qb_df.Position.fillna('QB')
```

```python
qb_df.head(10)
```

```python
#Renaming Column headers. Removing space in names
qb_df = qb_df.rename(columns = {'Passes Attempted': 'Passes_Attempted', 'Passes Completed':'Passes_Completed', 'Pass Attempts Per Game':'Pass_Attempts_Per_Game','TD Passes': 'TD_Passes','Passer Rating':'Passer_Rating' })
```

```python
qb_df.head(20)
```

```python
qb_df.tail(20)
```
