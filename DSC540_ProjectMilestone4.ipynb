```python
#Importing Libs
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#connection to Pro-Football Reference API
from bs4 import BeautifulSoup

url = 'https://www.pro-football-reference.com'
year = 2020

r = requests.get(url + '/years/' + str(year) + '/fantasy.htm')
soup = BeautifulSoup(r.content, 'html.parser')
parsed_table = soup.find_all('table')[0]  
```

```python
print("There are", format(len(parsed_table))," tables in total")
```

```python
#parsing table
parsed_table
```

```python
data = soup.findAll('tr')
```

```python
#length of data
print('Data is {} items long'.format(len(data)))
```

```python
#creating headers
headers = data[:2]
```

```python
headers
```

```python
#seperating data from headers
data_without_headers = data[3: ]
```

```python
data_without_headers
```

```python
col_headers = [[th.getText() for th in tr.findAll('th')] for tr in headers]
```

```python
col_headers
```

```python
#Editing Headers/ column names
col_headers2 =col_headers[1][1:]
```

```python
col_headers2 = ['Player',
 'Tm',
 'FantPos',
 'Age',
 'G',
 'GS',
 'Cmp',
 'PassAtt',
 'PassYds',
 'PassTD',
 'Int',
 'RushAtt',
 'RushYds',
 'Y/A',
 'RushTD',
 'Tgt',
 'Rec',
 'RecYds',
 'Y/R',
 'RecTD',
 'Fmb',
 'FL',
 'FanTD',
 '2PM',
 '2PP',
 'FantPt',
 'PPR',
 'DKPt',
 'FDPt',
 'VBD',
 'PosRank',
 'OvRank']
```

```python
#Collecting data
df_data= [[td.getText() for td in tr.findAll('td')] for tr in data_without_headers]
```

```python
#Creating Fantasy data frame
fandata = pd.DataFrame(df_data,columns = col_headers2)
```
