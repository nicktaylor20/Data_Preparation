Creating Football DB

```python
import sqlite3
from sqlite3 import Error


def create_connection(db_file):
    """ create a database connection to a SQLite database """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
    finally:
        if conn:
            conn.close()
   
```

```python
#Creating football database 
create_connection(r"C:\Users\nickm\Documents\football.db")
```

# Visualizations

```python
import matplotlib.pyplot as plt
```

```python
# set up the figure size
plt.rcParams['figure.figsize'] = (12, 10)
```

```python
c.execute("SELECT Year, SUM(Passes_Attempted) AS total_passes FROM qb_hist WHERE year BETWEEN 2010 and 2020 GROUP BY Year")
data = c.fetchall()
years = []
attempts = []

for row in data:
    years.append(row[0])
    attempts.append(row[1])
#plt.scatter(attempt)
fig, ax = plt.subplots()
ax.plot(attempts, years)

ax.set(xlabel='Pass Attempt ', ylabel='Year',
       title='Total Pass Attempts by Year ')
ax.grid()

plt.show()
```

```python
c.execute("SELECT Year, SUM(Rushing_Attempts) AS total_rushes FROM rb_hist WHERE year BETWEEN 2010 and 2020 GROUP BY Year")
data2 = c.fetchall()
years2 = []
attempts2 = []

for row in data2:
    years2.append(row[0])
    attempts2.append(row[1])
#plt.scatter(attempt)
fig, ax = plt.subplots()
ax.plot(attempts2, years2)

ax.set(xlabel='Rushing Attempts ', ylabel='Year',
       title='Total Rushing Attempts by Year ')
ax.grid()

plt.show()
```

```python
c.execute("SELECT FantPos, SUM(FantPt) FROM Fantasy_qb GROUP BY FantPos UNION ALL SELECT FantPos, SUM(FantPt) FROM Fantasy_rb GROUP BY FantPos")
data3 = c.fetchall()
pos = []
points = []

for row in data3:
    pos.append(row[0])
    points.append(row[1])


plt.bar(pos, points)
plt.title('Who Scores More Fantasy Points? QB vs RB', fontsize=15)
plt.ylabel('Total Fantasy Points', fontsize=12)
plt.xticks([0, 1], ['QBs', 'RBs'])
plt.tick_params(axis='both', labelsize=12)
plt.show()
```

```python
c.execute("SELECT * FROM (SELECT Player, Yds FROM qb_playoff LIMIT 5) UNION ALL  SELECT* FROM (SELECT Player, Yds FROM rb_playoff LIMIT 5) ")
data4 = c.fetchall()
name = []
yard = []


for row in data4:
    name.append(row[0])
    yard.append(row[1])

qbnames = name[0:5]
qbyards = yard[0:5]
rbnames = name[5:]
rbyards = yard[5:]

fig, axes = plt.subplots(nrows = 1, ncols = 2,figsize=(20, 8))
axes[0].bar(qbnames, qbyards)
axes[0].set_title('Top Playoff QBs', fontsize=20)
axes[0].set_ylabel('Passing Yards', fontsize=15)
axes[0].tick_params(axis='both', labelsize=12)
#axes[0].invert_xaxis()

axes[1].bar(rbnames, rbyards)
axes[1].set_title('Top Playoff RBs', fontsize=20)
axes[1].set_ylabel('Rushing Yards', fontsize=15)
axes[1].tick_params(axis='both', labelsize=12)

plt.show()
```

```python
c.execute("SELECT Name, SUM(Rushing_Yards_Per_Game) AS rushper FROM rb_hist GROUP BY Name ORDER BY rushper DESC LIMIT 5")

data5 = c.fetchall()
rbn = []
ypg = []

for row in data5:
    rbn.append(row[0])
    ypg.append(row[1])
    
fig, ax = plt.subplots()
ax.barh(rbn, ypg,align='center')
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Yards per Game')
ax.set_title('Rushing Yards Per Game Leaders',fontsize=20)
ax.tick_params(axis='both', labelsize=12)

plt.show()
```
