```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
#importing Boston housing data
df = pd.read_csv(r'C:\Users\nickm\Documents\Boston_housing.csv')
```

```python
#checking the first 10 records
df.head(10)
```

```python
#Total number of records
df.shape
```

```python
#Create a smaller new data frame
new_df = df[['CRIM', 'ZN', 'INDUS', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'PRICE']]
```

```python
#check last 7 records
new_df.tail(7)
```

```python
#plot a histogram for all variables 
import matplotlib.pyplot as plt
%matplotlib inline

for variable in new_df.columns:
    plt.hist(new_df[variable])
    plt.title(variable+' Histogram')
    plt.xlabel('Values')
    plt.ylabel('Distribution Count')
    plt.show()
    
```

```python
#creating a scatter plot
new_df.plot.scatter('CRIM', 'PRICE', s=50, c='red', edgecolor= 'k')
plt.title('Crime Rate vs Price')
plt.xlabel('Crime Rate')
plt.ylabel('Price')
plt.show()
```

```python
#Log of Crime scatter plot
plt.scatter(np.log10(new_df['CRIM']), new_df['PRICE'], s=50, c='red', edgecolor= 'k')
plt.title('Crime Rate Log10 vs Price')
plt.xlabel('Log Crime Rate')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
print('Mean rooms per dwelling ', new_df['RM'].mean())
print('Median Age of home', new_df['AGE'].median())
print('Mean distance to employment centers', new_df['DIS'].mean())
```
