```python
#importing BeautifulSoup lib
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
source = requests.get('https://www.pro-football-reference.com/leaders/pass_yds_career_playoffs.htm').text
soup = BeautifulSoup(source)
```

```python
print(soup.prettify())
```

```python
#finding the total number of tables
souptables= soup.find_all('table')
```

```python
print("There are", format(len(souptables))," tables in total")
```

```python
table = soup.table
print(table.contents)
```

```python
data = soup.findAll('tr')
```

```python
#length of data
print('Data is {} items long'.format(len(data)))
```

```python
data
```

```python
#creating headers
headers = data[0]
```

```python
headers
```

```python
#seperating data from headers
data_without_headers = data[1: ]
```

```python
data_without_headers
```

```python
col_headers = [th.getText() for th in headers.findAll('th')]
```

```python
#creating columns
col_headers
```

```python
#seperating data
df_data= [[td.getText() for td in tr.findAll('td')] for tr in data_without_headers]
```

```python
df_data
```

```python
#removing "rank" from column header
col_headers2 = col_headers[1: ]
```

```python
col_headers2
```

```python
qb_playoff_passing = pd.DataFrame(df_data, columns = col_headers2)
```
